#### 6.2 없는 데이터 nan 처리하기
- nan : csv 에서 빈셀을 의미 . 값이 없다

빈 값이 있는지 확인하기
- isnull : 빈값이 있는지 확인
- isnuull.sum : 빈값의 개수

```python
# 서울 마포구 코로나 현황
data = pd.read_csv('0917/서울특별시 마포구_법정동별_월별_코로나19 확진자 현황.csv', encoding='euc=kr')
data
df = data [['기준연도','기준월','법정동','확진자수']]
df
# 확진자수에 빈값이 있는지 확인
df['확진자수'].isnull()
# 빈값 개수 세기
# bool -> int
# false : 0
# true : 1
df['확진자수'].isnull().sum()
```

반값 삭제하기

- dropna : 빈값이 있는 행을 삭제
- dropna(subset=['컬럼'] = 빈값이 있는지 특정 컬럼만 검사하여 삭제

```python
# 빈값이 있는 행을 모두 삭제
df.dropna()
# 확진자수가 없는 행만 삭제
df.dropna(subset=['확진자수'])
# 원본
df
```

빈값을 다른값으로 대체하기
- fillna(새로운값): 컬럼 하나만 처리할때
- fillna({컬럼1:새로운값,컬럼2:새로운값}): 여러컬럼을 한번에 처리할때

```python
# 변경된 내용을 원본에 넣기
#df['법정동'] = df['법정동'].fillna('모름')
#df
#df['확진자수'] = df['확진자수'].fillna(0)
#df
#법정동에 빈값이 있으면 '모름' 으로 표시
#df['법정동'].fillna('모름')
# 확진자수에 빈값이 있으면 0으로 표시
#df['확진자수'].fillna(0)
# -> 원본은 그대로
# 여러컬럼을 한번에 교체
# 법정동과 확진자수의 빈값을 한번에 처리
df.fillna({'법정동':'모름' ,'확진자수':0})
df
```
